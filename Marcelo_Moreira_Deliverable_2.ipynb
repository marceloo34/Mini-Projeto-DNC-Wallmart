## Importanto Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Lendo DataFrame

```python
df = pd.read_csv('Walmart.csv')
df
```

```python
df.info()
```

## Formatando Valores

```python
# Alterar tipo do dado de 'Date' para formato data.
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

## Correlações

```python
tabela_correlacoes = df.groupby(['Store']).agg('corr').reset_index()
tabela_correlacoes = tabela_correlacoes[tabela_correlacoes['Weekly_Sales'] == 1].reset_index()
tabela_correlacoes = tabela_correlacoes.drop(columns=['level_1', 'index'])
tabela_correlacoes
```

```python
correlacao_total = tabela_correlacoes.drop(columns=['Store', 'Weekly_Sales'])
figura = plt.figure(figsize = (20, 40))
sns.heatmap(correlacao_total, annot=True)
plt.show()
```

## Tomando Decisões com as Correlações

```python
tabela_correlacoes.describe(percentiles=[.05, .5, .95])
```

```python
tabela_maiores_corr = tabela_correlacoes[(tabela_correlacoes['Holiday_Flag'] >= 0.24) |
                                         ((tabela_correlacoes['Temperature'] >= 0.17) | (tabela_correlacoes['Temperature'] <= -0.378)) |
                                         ((tabela_correlacoes['Fuel_Price'] >= 0.32) | (tabela_correlacoes['Fuel_Price'] <= -0.416)) |
                                         ((tabela_correlacoes['CPI'] >= 0.42) | (tabela_correlacoes['CPI'] <= -0.396)) |
                                         (tabela_correlacoes['Unemployment'] <= -0.38)
                                         ].reset_index()
tabela_maiores_corr
```

```python
figura = plt.figure(figsize = (10, 5))
g = sns.barplot(data = tabela_maiores_corr, x = 'Store', y = 'Holiday_Flag')
g.set_title('CORRELAÇÃO POR LOJA: Vendas x Feriado')
plt.show()
```

```python
figura = plt.figure(figsize = (10, 5))
g = sns.barplot(data = tabela_maiores_corr, x = 'Store', y = 'Temperature')
g.set_title('CORRELAÇÃO POR LOJA: Vendas x Temperatura')
plt.show()
```

```python
figura = plt.figure(figsize = (10, 5))
g = sns.barplot(data = tabela_maiores_corr, x = 'Store', y = 'Fuel_Price')
g.set_title('CORRELAÇÃO POR LOJA: Vendas x Combustível')
plt.show()
```

```python
figura = plt.figure(figsize = (10, 5))
g = sns.barplot(data = tabela_maiores_corr, x = 'Store', y = 'CPI')
g.set_title('CORRELAÇÃO POR LOJA: Vendas x IPC')
plt.show()
```

```python
figura = plt.figure(figsize = (10, 5))
g = sns.barplot(data = tabela_maiores_corr, x = 'Store', y = 'Unemployment')
g.set_title('CORRELAÇÃO POR LOJA: Vendas x Desemprego')
plt.show()
```
